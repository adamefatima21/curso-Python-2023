# crear un flujo con condicionales

```python
import pandas as pd
pd.DataFrame(columns=['Ingresos', 'Tasa impositiva'],
             index=[1, 2, 3, 4],
             data=[['Menos de 1,000', '5 %'],
                   ['Entre 1,000 y 2,000', '15 %'],
                   ['Entre 2,000 y 3,000', '20 %'],
                   ['Más de 3,000', '30 %']])
```

```python
ingreso_usuario = float(input("cual es tu ingreso mensual? "))
if ingreso_usuario > 0 and ingreso_usuario <= 1000:
  tasa = 0.05

elif ingreso_usuario >1000 and ingreso_usuario <= 2000:
  tasa = 0.15

elif ingreso_usuario >2000 and ingreso_usuario <= 3000:
  tasa = 0.2

elif ingreso_usuario>3000:
  tasa = 0.3

else:
  print("Deberias ingresar un valor mayor a 0 ")

impuestos = ingreso_usuario * tasa
print("tus impuestos a pagar son: $", impuestos)
```
